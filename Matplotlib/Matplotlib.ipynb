No mtodo fig.add_axes(), os argumentos [left, bottom, width, height] no so posies relativas, mas sim coordenadas absolutas que especificam a posio e o tamanho dos eixos dentro da figura.

left: a posio horizontal esquerda da caixa de eixos na figura, em coordenadas da figura (0   esquerda).
bottom: a posio vertical inferior da caixa de eixos na figura, em coordenadas da figura (0  na parte inferior).
width: a largura da caixa de eixos em coordenadas da figura.
height: a altura da caixa de eixos em coordenadas da figura.

```python
#2nd method
fig  = plt.figure()#Create a figure
ax = fig.add_axes([1,1,1,1,1,])
ax.plot(x,y)
plt.show()
```

```python
#Matplot example worklflow
# 0-> Import matplotlib and get it ready for plotting in jupyeter

import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10,10))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title ="Simple Plot",
       xlabel ="x-axis",
       ylabel = "y-axis")
#5. Save & show (you save the whole figure)
fig.savefig("Imagens/sample-plot.png")
```

### Which one should you use? (pyplot vs matplotlib OO method)
 * When plotting something quicly, okay to uso the pyplot method
 * When plotting something more advanced, use the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
            figsize = (10,6));
```

```python
##OO method mixed with pyplotmnethod
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax)
#ax.set_xlim([45,100])#Colocando um intervalo para idade
```

```python
##00method from scratch
fig,ax = plt.subplots(figsize = (10,6))

#plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

#Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "Cholesterol");

#add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle = "--");
```

```python
over_50.head()
```

```python
#Subplot of chol,age thalach 
fig,(ax0,ax1) = plt.subplots(nrows =2,
                             ncols=1,
                             figsize = (10,10));

#add data to ax0
sacatter = ax0.scatter(x=over_50["age"],
                       y=over_50["chol"],
                       c=over_50["target"])
#Customiz3e ax0
ax0.set(title ="Heart Disease and Cholesterol Levels",
        xlabel = "Age",
        ylabel ="Cholesterol");

##add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

##add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");

##Add data to axl
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])
#Customize ax1
ax1.set(title="Heart Diesease and Max Heart Rate",
        xlabel = "Age",
        ylabel ="Max Heart Rate");
#Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title = "Target")
#aad a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle ="--");


##ADd a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize = 16,fontweight="bold");
```
